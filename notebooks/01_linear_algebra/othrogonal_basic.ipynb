```python
import numpy as np

# 1. ベクトルを定義
a = np.array([3, 4])  # 投影されるベクトル (斜めの棒)
b = np.array([4, 0])  # 投影先のベクトル (地面)

print("a", a)
print("b", b)

# 2. 内積 (a・b) を計算
dot_product = np.dot(a, b)

print("a・b", dot_product)

# 3. bの長さの2乗 (|b|^2) を計算
# ノルムの2乗は、自分自身との内積と同じです (b・b)
b_norm_sq = np.dot(b, b)
# b_norm_sq = np.sum(b**2)
# b_norm_sq = np.linalg.norm(b)**2
print("|b|^2", b_norm_sq)

# 4. 正射影ベクトル p を計算
# 公式: (a・b / |b|^2) * b
p = (dot_product / b_norm_sq) * b

print(f"ベクトル a: {a}")
print(f"ベクトル b: {b}")
print(f"正射影ベクトル p: {p}")

# この例では、$(3, 4)$ というベクトルをX軸上の $(4, 0)$ に投影したので、Y成分（高さ）が消えて $(3, 0)$ という影ができました。
```

```python
import numpy as np
import matplotlib.pyplot as plt

# ベクトル定義
a = np.array([2, 3])
b = np.array([4, 1])

# 正射影の計算
dot_product = np.dot(a, b)
b_norm_sq = np.dot(b, b)
p = (dot_product / b_norm_sq) * b

# --- 以下は描画コード ---
fig, ax = plt.subplots()

# 原点 (0,0)
origin = [0], [0]

# ベクトルを描画 (quiverプロットを使用)
ax.quiver(*origin, a[0], a[1], color='r', angles='xy', scale_units='xy', scale=1, label='Vec a (Target)')
ax.quiver(*origin, b[0], b[1], color='b', angles='xy', scale_units='xy', scale=1, label='Vec b (Base)')
ax.quiver(*origin, p[0], p[1], color='g', angles='xy', scale_units='xy', scale=1, label='Vec p (Projection)', width=0.015)

# 補助線（aの先端からpの先端へ点線を引く＝垂直な光）
plt.plot([a[0], p[0]], [a[1], p[1]], 'k--', alpha=0.5, label='Perpendicular Line')

# グラフの設定
plt.xlim(-1, 5)
plt.ylim(-1, 5)
plt.grid()
plt.legend()
plt.gca().set_aspect('equal', adjustable='box') # アスペクト比を1:1にして歪みを防ぐ
plt.title('Orthogonal Projection Vector')

plt.show()
```
