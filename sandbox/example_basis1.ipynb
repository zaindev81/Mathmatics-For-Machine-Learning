```python

import numpy as np
```

```python
standard_basis= np.array([[1,0], [0,1]])


print("standard_basis", standard_basis)
print("standard_basis", standard_basis.shape)
print("standard_basis", standard_basis.ndim)

print("np", np.linalg.matrix_rank(standard_basis))
print("len", len(standard_basis))

another_basis = np.array([[1,1], [-1, 2]])

print("another_basis", another_basis)
print("another_basis", another_basis.shape)
print("another_basis", another_basis.ndim)

print("np", np.linalg.matrix_rank(another_basis))
print("len", len(another_basis))



```
