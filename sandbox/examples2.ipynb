```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Basic
standard_basis = np.array([
    [1, 0], # e1
    [0, 1]  # e2
])

print("standard_basis", standard_basis)
print("rank", np.linalg.matrix_rank(standard_basis))
print("len(standard_basis)", len(standard_basis))
print("Linear Independence", np.linalg.matrix_rank(standard_basis) == len(standard_basis))


# Linearly Dependent
spanning_set = np.array([
    [1, 0], # e1
    [0, 1], # e2
    [1, 1] # e1 + e2
])

print("spanning_set", spanning_set)
print("rank", np.linalg.matrix_rank(spanning_set))
print("len(spanning_set)", len(spanning_set))
print("Linearly Dependent", np.linalg.matrix_rank(spanning_set) < len(spanning_set))
```

```python
# Standard Basis

v = np.array([3, 4])
print("v", v)

e1 = np.array([1,0])
e2 = np.array([0,1])

print("v", v)
print("e1", e1)
print("e2", e2)

print("v = 3e1 + 4e2", 3*e1 + 4*e2)

fig, ax = plt.subplots(1, 1, figsize=(8, 8))

ax.arrow(0, 0, e1[0], e1[1], head_width=0.2, head_length=0.2,
         fc='blue', ec='blue', linewidth=2, label='e1 (standard basis)')
ax.arrow(0, 0, e2[0], e2[1], head_width=0.2, head_length=0.2,
         fc='green', ec='green', linewidth=2, label='e2 (standard basis)')

ax.arrow(0, 0, v[0], v[1], head_width=0.2, head_length=0.2,
         fc='red', ec='red', linewidth=2, label=f'v = {v}')

ax.grid(True, alpha=0.3)
ax.set_xlim(-1, 5)                 # X軸の表示範囲を-1から5に設定
ax.set_ylim(-1, 5)                 # Y軸の表示範囲を-1から5に設定
ax.set_aspect('equal')             # 縦横比を1:1にする（これをしないと矢印が歪んで見えます）

ax.axhline(y=0, color='k', linewidth=0.5) # X軸（y=0の線）を黒色で引く
ax.axvline(x=0, color='k', linewidth=0.5) # X軸（y=0の線）を黒色で引く

ax.legend()                        # 各矢印のラベル（凡例）を表示
ax.set_title('Vector Representation in Standard Basis', fontsize=14)
ax.set_xlabel('x')
ax.set_ylabel('y')

plt.tight_layout()
plt.show()
```
